```python
graph = {
    0: [1, 2, 5],
    1: [0, 2, 3, 4],
    2: [0, 1, 3, 5],
    3: [1, 2, 4, 6],
    4: [1, 3, 6, 7],
    5: [0, 2, 7, 8],
    6: [3, 4, 7, 9],
    7: [4, 5, 6, 9],
    8: [5, 9],
    9: [6, 7, 8]
}
```

```python
import networkx as nx
import matplotlib.pyplot as plt

def greedy_coloring(graph):
    coloring = {}
    
    for vertex in graph:
        #find color of neighbors vertices
        adjacent_colors = {coloring[neighbor] for neighbor in graph[vertex] if neighbor in coloring}
        
        #the smallest available color
        color = 0
        while color in adjacent_colors:
            color += 1
        coloring[vertex] = color
    return coloring

def plot_greedy_coloring(graph, coloring):
    G = nx.Graph()
    
    #Add nodes and edges
    for vertex, neighbors in graph.items():
        for neighbor in neighbors:
            G.add_edge(vertex, neighbor)
    
    #color mapping
    colors = [coloring[node] for node in G.nodes]
    
    plt.figure(figsize=(8, 6))
    nx.draw(
        G,
        with_labels=True,
        node_color=colors,
        node_size=800,
        cmap=plt.cm.get_cmap('tab20'),
        font_weight='bold'
    )
    plt.title("Graph Coloring using Greedy Algorithm")
    plt.show()

coloring_result = greedy_coloring(graph)

print("Vertex Coloring:")
for vertex, color in coloring_result.items():
    print(f"Vertex {vertex}: Color {color}")

plot_greedy_coloring(graph, coloring_result)
###number coloring must be added
```

```python
def number_coloring(graph):
    numbering = {}  

    for vertex in graph:
        # Get the numbers already assigned to adjacent vertices
        adjacent_numbers = {numbering[neighbor] for neighbor in graph[vertex] if neighbor in numbering}

        number = 1
        while number in adjacent_numbers:
            number += 1
        
        numbering[vertex] = number

    return numbering

number_coloring_result = number_coloring(graph)

print("Vertex Numbering:")
for vertex, number in number_coloring_result.items():
    print(f"Vertex {vertex}: Number {number}")

def plot_number_coloring(graph, numbering):
    G = nx.Graph()

    # Add edges to the graph
    for vertex, neighbors in graph.items():
        for neighbor in neighbors:
            G.add_edge(vertex, neighbor)

    # Get numbers assigned to each vertex
    numbers = [numbering[node] for node in G.nodes]

    # Plot the graph
    plt.figure(figsize=(8, 6))
    nx.draw(
        G,
        with_labels=True,
        node_color=numbers,  # Node colors based on assigned numbers
        node_size=800,
        cmap=plt.cm.get_cmap('tab20'),
        font_weight='bold'
    )
    plt.title("Graph Numbering using Number Coloring Algorithm")
    plt.show()

# Visualize the numbered graph
plot_number_coloring(graph, number_coloring_result)
```
